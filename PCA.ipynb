' PCA is a linear transformation that changes the coordinate system to orthogonal axes ordered by variance.\nThe first principal component captures the maximum variance in the data, and each subsequent component captures the maximum remaining variance perpendicular to the previous ones.\nIn many practical datasets, later components often correspond to noise.'

```python
import numpy as np
import matplotlib.pyplot as plt
```

What PCA solves? 

PCA essentially solves the problem of plotting data with 4 or more variable( 4D or more) This essentially creates a 2D or a 3D plot of axes which capture the direction of highest subsequent Variances (Principle component)  

```python
np.random.seed(0)
X = np.random.randn(200, 2 ) # create a 2D array with random values. 
# currently there is no correlation just randomness. 
#print(X)
# lets make the column 1 depend on column 2 
X[: , 1] = np.random.randint(1,29)* X[:, 0 ] + 10 * np.random.randn(200)


mean = X.mean(axis= 0)
# calculate feature-wise mean essentially using broadcasting this vector will help u scale the vector to many dimensions
X = X-mean
plt.scatter(X[:,0], X[:,1] )
plt.title("Random plot for PCA")
plt.xlabel('x')
plt.ylabel('y') 
plt.show()
```

The above code cell create a 2D array of 2Variables (x and y) x is just pure random numbers and y shows a relationship and a small noise component
y = mx + noise 

```python
covariance_matrix = np.cov(X, rowvar = 0)
eigen_value , [eigen_vector1, eigen_vector2 ] = np.linalg.eig(covariance_matrix)
print(eigen_value)
print(eigen_vector2)
print(eigen_vector1)
```

```python
pc1 = np.dot(X, eigen_vector2)
pc2 = np.dot(X, eigen_vector1)
print(np.var(pc1))
print(np.var(pc2))
plt.scatter(pc1,pc2)
plt.xlabel('PC1')
plt.ylabel('PC2')
```
